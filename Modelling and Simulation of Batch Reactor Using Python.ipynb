```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#Simulation Parameter
t_start = 0
t_End = 100
initial_concentration_A = 1.0
initial_concentration_B = 0.5
initial_concentration_C = 0
rate_constant            = 0.05
```

```python
#Defining the system of differential system
def reactionODEs(y,t):
    A,B,C = y
    
    dA_dt = -rate_constant*A*B
    dB_dt = -rate_constant*A*B
    dC_dt = rate_constant*A*B
    return[dA_dt,dB_dt,dC_dt]
```

```python
#Setting up initial condition
initial_conditions = [initial_concentration_A,initial_concentration_B,initial_concentration_C]
```

```python
#Time point for the simulation
t = np.linspace(t_start,t_End,500)
```

```python
#Solving the ODEs numerically
concentration = odeint(reactionODEs,initial_conditions,t)
```

```python
#Extract the concentraction of  A,B and C
#Concentration Array
concentration_A = concentration[:, 0]
concentration_B = concentration[:, 1]
concentration_C = concentration[:, 2]
```

```python
# Plot results
plt.plot(t, concentration_A,'r', label='[A]')
plt.plot(t, concentration_B,'b', label='[B]')
plt.plot(t, concentration_C,'g', label='[C]')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.title('Batch Reactor Simulation with Second Order Reaction')
plt.grid()
plt.show()
```
